# **8. Exploratory Data Analysis**
---



## Analysis based on number of films released year-wise.

```python
data['Year'].value_counts(sort=False)
```

```python
plt.figure(figsize = (10,5))
data['Year'].value_counts(sort = False).plot(kind='bar', color='g')
plt.ylabel('Number of Movies')
```

- We notice a **slight drop** in number of movies **in 2009 compared to 2008**.

- Thereafter we consistently see increase in number of movies, with number of movies in 2016 almost double of those in 2015.

```python
data_yrev = data.groupby(['Year'])['Revenue (Millions)'].sum()
data_yrev
```

```python
plt.figure(figsize = (10,5))
data_yrev.plot(kind = 'bar', figsize=(8,5), fontsize = 10, color = 'green')
plt.ylabel('Revenue in Million $')
```

- In the previous graph, we saw slight drop in 2009 compared to 2008. But **revenue-wise**, 2009 has outperformed 2008 by **11.26%**.

- Compared to **2010, 2011** saw a **drop in revenue** to the tune of **~11%**.

## Top 4 directors by number of movies directed.

```python
data['Director'].value_counts()
```

```python
options = ['Ridley Scott', 'David Yates', 'Michael Bay', 'Paul W.S. Anderson']
data_dir = data[data['Director'].isin(options)]
data_dir
```

```python
plt.figure(figsize = (8,5))
data_dir['Director'].value_counts().plot(kind = 'bar', color = 'b')
plt.ylabel('Number of Movies Directed')
```

- Out of the **total 838 movies, top 4 directors account for 26 movies**. That is **3.1%** of all movies.

    - Among the above **26 movies**, **15 movies**  belong to **Action** genre.
    
    - Also **17 movies** belong to **Adventure** genre.

- Among top 4 directors, **Ridley Scott has directed the maximum - 8**

    - That is 25% more than the second placed directors.

## Top Directors by Revenue

```python
data['Revenue (Millions)'].sum()
```

```python
data_rev = data.groupby(['Director'])['Revenue (Millions)'].sum().sort_values(ascending=False)
data_rev.head()
```

```python
data_rev.head().sum()
```

```python
data_rev.head().plot(kind = 'bar', figsize=(8,5), fontsize = 10, color = 'green')
plt.ylabel('Total Revenue (in Million $)')
```

- We notice that **David Yates** and **Michael Bay**, who feature in **Top director movie-wise** also feature here.

- These **5 directors** account for **7550.18 million dollars** out of total **70865 million dollars** i.e. nearly **10.65%** of total revenue.

```python
options = ['J.J. Abrams', 'David Yates', 'Christopher Nolan', 'Michael Bay', 'Francis Lawrence']
data_dir = data[data['Director'].isin(options)]
data_dir
```

- The **top 5 directors by revenue** have directed **26 movies** amongst themselves.
 
- Of these 26 movies -
 
     - **19 movies** belong to **Adventure** genre.
     
     - **16 movies** belong to **Action** genre.
     
     - **13 movies** belong to **Sci-Fi** genre.

```python
data['Runtime (Minutes)'].value_counts(bins=5).plot(kind='pie', fontsize=14, autopct='%2.1f%%', figsize=(9,9), legend=True, cmap='summer')
plt.legend(loc='upper right', fontsize='small')
plt.title('Pie-chart showing Movie Runtime')
```

- **Runtime** bucketed into **5 bins**.

- Majority (**48.8%**) of the films have **Runtime between 90 mins and 114 mins**

- Very less (**1.9%**) movies stretch beyond **2.5 hours of runtime**.

### **9.2 Actionable Insights**

Some of the points to be considered inorder to bring crowd back to theatre - 
- Better to make an **Action** or an **Adventure** movie, as they give better ROI.

- Rather than making **independent stand-alone movies**, focus should be given to **multi-part movies**. This ensures in return   of loyal fans, which results good box-office returns.

- Priority should be on movies with **duration between 1.5 hrs - 2 hrs**. People don't prefer to sit for **long-duration**.

- Also **short duration movies means more shows per day**, which results good box-office returns.
